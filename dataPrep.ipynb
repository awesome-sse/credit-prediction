```python
def toLower(s):
    return s.replace(' ', '').lower()

def appPotrebEncode(x):
    if x == 0:
        return 0
    return 1
```

```python
import pandas as pd
import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = './data/'
filename = 'aggregates'

df = pd.read_csv(path + filename + '.csv')
#app_potreb_cnt
```

```python
df = df.drop(df[df.app_potreb_cnt.isnull()].index)

to_drop = ['report_date', 'client_id']
df.drop(to_drop, axis=1, inplace=True)
df = df.loc[:, (df.isnull().sum(axis=0)/len(df.app_potreb_cnt) <= 0.3)]
```

```python
for p in df.select_dtypes(include=['object']):
    df[p] = [toLower(str(x)) for x in df[p]]
    
for p in df.select_dtypes(include=['object']):
    names = list(set(df[p]))
    df[p] = [names.index(x) for x in df[p]]
```
